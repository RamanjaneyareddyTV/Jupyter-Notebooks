<h3> <center>  Billboard  </center> </h3>

```python
billboard = pd.read_csv("Files/billboard.csv", encoding='mac_latin2')
billboard.head()
```

```python
# Melt function
id_vars = ["year","artist.inverted","track","time","genre","date.entered","date.peaked"]
billboard = pd.melt(frame=billboard, id_vars=id_vars, var_name="week", value_name="rank")

# Drop NaN values
billboard = billboard.dropna()

# Change type
billboard["week"] = billboard['week'].str.extract('(\d+)', expand=False).astype(int)
billboard["rank"] = billboard["rank"].astype(int)

billboard['date'] = pd.to_datetime(billboard['date.entered']) + pd.to_timedelta(billboard['week'], unit='w') - pd.DateOffset(weeks=1)
#billboard = billboard.drop('year', axis=1)

billboard = billboard[["year", "artist.inverted", "track", "time", "genre", "week", "rank", "date"]]
billboard = billboard.sort_values(ascending=True, by=["year","artist.inverted","track","week","rank"])
billboard.reset_index(inplace=True)
billboard = billboard.drop(billboard.columns[[0, 1]], axis=1)
billboard.head()
# Done
```

<h3> <center>  Tuberculosis  </center> </h3>

```python
tb = pd.read_csv("Files/tb.csv")
tb.head()
```

```python
# Melt
tb = pd.melt(tb, id_vars=["iso2","year"], value_name="cases", var_name="sex_and_age")

# Extract Sex, Age lower bound and Age upper bound group
tb_temp = tb["sex_and_age"].str.extract("(\D)(\d+)(\d{2})", expand=False)    

# Name columns
tb_temp.columns = ["sex", "age_lower", "age_upper"]

# Create `age`column based on `age_lower` and `age_upper`
tb_temp["age"] = tb_temp["age_lower"] + "-" + tb_temp["age_upper"]

# Merge 
tb = pd.concat([tb, tb_temp], axis=1)

# clean the table
tb = tb.drop(['sex_and_age',"age_lower","age_upper"], axis=1)
tb = tb.dropna()
tb.rename(columns={'iso2':'country'}, inplace=True)
tb = tb.sort_values(ascending=True,by=["country", "year", "sex", "age"])
tb.reset_index(inplace=True)

tb.head(10)
```

<h3> <center>  Weather  </center> </h3>

```python
weather = pd.read_csv("Files/weather.txt", sep="\t")
weather.head()
```

```python
# Melt
weather = pd.melt(weather, id_vars=["id", "year","month","element"], var_name="day")

# CLean the table
weather.dropna(inplace=True)
weather = weather.reset_index(drop=True)

# Format for the 'date' table
def f(row):    
    return "%d-%02d-%02d" % (row['year'], row['month'], int(row['day'][1:]))
weather['date'] = weather.apply(f,axis=1)

# configure the columns
weather = weather[['id', 'element','value','date']]
weather = weather.pivot(index='date',columns='element',values='value')
weather['id'] = 'MX17004'
weather.reset_index(inplace=True)
weather.columns.name = ''

weather.head()
```
