```python
import pandas as pd
import numpy as np
```

```python
# NFL Games
nfl = pd.read_csv("Files/NFL Play by Play 2009-2017 (v4).csv")

# San Francisco Building permits
af_permits = pd.read_csv("FIles/Building_Permits.csv")
```

<center> <h2> NFL <h2/> <center/>

```python
nfl.info()
```

```python
# Dataset shape
nfl_shape = nfl.shape
nfl_shape
```

```python
# Missing Values (obtain the column with most missing values)
miss_values = nfl.isnull().sum().sort_values(ascending=False)
miss_values.head()
```

```python
# Compute: Missing values cells/ total cells
total_cells = np.product(nfl_shape)
print("Total cells:", total_cells)

# Total missing
total_missing = miss_values.sum()
print("Total Missing Values:", total_missing)

# Percentage
percentage = total_missing / total_cells*100
print("Percentage:", percentage.round())
```

```python
# Deleting all rows
nfl.dropna()
```

```python
# Deleting all columns with Missing Values
nfl_deleted = nfl.dropna(axis=1)
nfl_deleted.shape
```

<center> <h2> Building Permits <h2/> <center/>

```python
af_shape = af_permits.shape
af_shape
```

```python
miss_values_bp = af_permits.isnull().sum().sort_values(ascending=False)
miss_values_bp.head()
```

```python
# Compute: Missing values cells/ total cells
total_cells = np.product(af_shape)
print("Total cells:", total_cells)

# Total missing
total_missing = miss_values_bp.sum()
print("Total Missing Values:", total_missing)

# Percentage
percentage = total_missing / total_cells*100
print("Percentage of missing values:", percentage.round())
```
